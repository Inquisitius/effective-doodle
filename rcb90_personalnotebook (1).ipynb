The submetric I selected was Covid-19 Infection Rates because they demonstrate how well specific neighborhoods 
responded to this specific pandemic. 

https://data.wprdc.org/dataset/allegheny-county-covid-19-tests-cases-and-deaths/resource/0f214885-ff3e-44e1-9963-e9e9062a04d1

Because the data from Covid-19 shows us how each neighborhood responded to the virus, we're able to develop an understanding of how well these Pittsburgh neighborhoods would respond when faced with an even worse virus, if the events of The Last of Us ever occured.

```python
#Setting up the data

import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = 'covid-19.csv' 

# Input the file path

df  = pd.read_csv(csv_file_path)

#Slicing and pivoting the data so that it fits the set

pivot_df = df.pivot(columns='neighborhood_municipality', values='infections')

#Sorting the data so that we can simplfiy the number of neighborhoods we need to observe

sorted_df = pivot_df.sum().sort_values(ascending=False)

# Select the lowest 20 neighborhoods and plotting them

lowest_20 = sorted_df.tail(20)
lowest_20.plot(kind='bar', figsize=(12, 6), xlabel='Neighborhood', ylabel='Infections', title='Neighborhoods with Lowest 20 Infection Rates')
plt.show()
```
